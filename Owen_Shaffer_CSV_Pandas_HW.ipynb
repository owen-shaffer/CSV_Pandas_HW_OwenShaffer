```python
import pandas as pd

acc_players = pd.read_csv('acc_players-2324F.csv')

acc_players.head(5)
```

```python
acc_players.columns = acc_players.iloc[0]
acc_players = acc_players[1:]
```

```python
acc_players.head(5)
```

```python
acc_players['PTS'] = pd.to_numeric(acc_players['PTS'], errors='coerce')

total_points = acc_players['PTS'].sum()
print(total_points)

```

```python
acc_players['MP'] = pd.to_numeric(acc_players['MP'], errors='coerce')
top_player = acc_players.loc[acc_players['MP'].idxmax()]
print(top_player)
```

```python
acc_players['TRB'] = pd.to_numeric(acc_players['TRB'], errors='coerce')
top_rebounders = acc_players.sort_values('TRB',ascending=False).head(5)
print(top_rebounders[['Player', 'TRB']])
```

```python
acc_players['MP'] = pd.to_numeric(acc_players['MP'], errors='coerce')
acc_players['AST'] = pd.to_numeric(acc_players['AST'], errors='coerce')
acc_players['BLK'] = pd.to_numeric(acc_players['BLK'], errors='coerce')

playersover500 = acc_players[acc_players['MP'].astype(int) > 500]
playersover500.head(5)
```

```python
player_with_highest_assists = playersover500.loc[playersover500['AST'].astype(int).idxmax()]
print( player_with_highest_assists['Player'])
```

```python

top_3_assist_leaders = acc_players.sort_values('AST', ascending=False).head(3)
print(top_3_assist_leaders[['Player', 'AST']])
```

```python
top_3_shot_blockers = acc_players.sort_values('BLK', ascending=False).head(3)
print(top_3_shot_blockers[['Player', 'BLK']])
```

```python
total_points_by_school = acc_players.groupby('School')['PTS'].sum()
print(total_points_by_school)
```

```python
total_assists_by_school = acc_players.groupby('School')['AST'].sum()
print(total_assists_by_school)
```

```python
top_3_schools_by_points = total_points_by_school.sort_values(ascending=False).head(3)
print(top_3_schools_by_points)
```

```python
import matplotlib.pyplot as plt

top_5_scorers = acc_players.sort_values('PTS', ascending=False).head(5)
plt.figure(figsize=(6, 6))
plt.bar(top_5_scorers['Player'], top_5_scorers['PTS'])
plt.xlabel('Player')
plt.ylabel('Total Points')
plt.title('Top 5 Players by Total Points Scored')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
acc_players['FG%'] = pd.to_numeric(acc_players['FG%'], errors='coerce')
acc_players['WS'] = pd.to_numeric(acc_players['WS'], errors='coerce')

plt.figure(figsize=(6, 6))
plt.scatter(acc_players['FG%'], acc_players['WS'])
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.title('Correlation between FG% and WS')
plt.show()

correlation = acc_players['FG%'].corr(acc_players['WS'])
print(f"Correlation coefficient: {correlation}")
```
